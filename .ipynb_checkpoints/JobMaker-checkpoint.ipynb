```python
# molpro_job_template = jinja2.Template("""***,Molpro Input
# memory,480,m    ! memory: MEM_PER_CPU/8 * ~(0.90 to 0.95)
# symmetry,nosym
# orient,noorient
# geometry={
# angstrom;
# {{XYZ}}
# }

# basis={
# default=aug-cc-pvtz-f12

# set,jkfit,context=jkfit
# default=aug-cc-pvtz-f12

# set,mp2fit,context=mp2fit
# default=aug-cc-pvtz-f12
# }

# ! Set wave function properties
# wf,spin=0,charge=0

# ! F12 parameters
# explicit,ri_basis=jkfit,df_basis=mp2fit,df_basis_exch=jkfit

# ! density fitting parameters
# cfit,basis=mp2fit

# {df-hf,basis=jkfit}
# {df-mp2-f12,cabs_singles=-1}
# {pno-lccsd(t)-f12}
# """)
molpro_job_template = jinja2.Template("""***,Molpro Input
memory,480,m    ! memory: MEM_PER_CPU/8 * ~(0.90 to 0.95)
symmetry,nosym
orient,noorient
geometry={
angstrom;
{{XYZ}}
}

basis=avdz

! hf energy
hf

""")
```

```python
molpro_slurm_template = jinja2.Template("""#!/bin/bash
#SBATCH --job-name={{NAME}}
#SBATCH --partition=long
#SBATCH --nodes=1
#SBATCH --ntasks=4
#SBATCH --cpus-per-task=1
#SBATCH --mem-per-cpu=6000

#---------
# Modules
#---------

# PC-Beethoven
#module molpro/molpro2022-gcc-9.2.0
# PC-Bach
module load molpro/molpro-2022.1

#------------
# Parameters
#------------

export MAINDIR=$(pwd)

# Molpro file (expect input file with .inp suffix, but don't put here, see below)
export MOLFILE={{NAME}}  #your_molpro_file

export INPFILE=$MOLFILE.inp
export OUTFILE=$MOLFILE.out

# Make temporary directory
export TMPDIR=/scratch/$USER/jobs/tmp.$SLURM_JOBID
if [ -d $TMPDIR ]; then
  echo "$TMPDIR exists; double job start; exit"
  exit 1
fi
mkdir -p $TMPDIR

#-------------
# Computation
#-------------

# Run MOLPRO
molpro -d $TMPDIR -I $TMPDIR -W $TMPDIR -o $TMPDIR/$OUTFILE --no-xml-output $MAINDIR/$INPFILE

# Copy result file to output directory
cp $TMPDIR/$OUTFILE $MAINDIR/$OUTFILE

# Remove temporary directory
rm -rf $TMPDIR

""")
```

```python
pdbpath = "/home/boittier/charmm/test/pdbs4"
z2a = {1:"H", 8: "O"}

def make_dir(path):
    """Make a directory if it doesn't exist"""
    if not os.path.exists(path):
        os.makedirs(path)

```

```python
def make_jobs(i, 
              job_dir="/home/boittier/pcbach/3body/hf/"):
    key = cluster_flux["key"].iloc[i]
    atoms = read(os.path.join(pdbpath, key+".xyz.pdb"))
    strs = []
    for a,xyz in zip(atoms.numbers,atoms.positions):
        strs.append(f"{z2a[a]} {xyz[0]} {xyz[1]} {xyz[2]}\n")
    XYZ = "".join(strs)
    
    job_str = molpro_job_template.render(XYZ=XYZ)
    make_dir(job_dir+key)
    with open(job_dir+key+f"/{key}.inp", "w") as f:
        f.write(job_str)
    with open(f"{job_dir}/{key}"+f"/{key}.sh", "w") as f:
        f.write(molpro_slurm_template.render(NAME=key))
    _key = key
    
    for i in range(len(atoms)):
        key = _key+"_"+str(i)
        job_str = molpro_job_template.render(XYZ="\n".join(XYZ.split("\n")[i*3:i*3+3]))
        with open(job_dir+_key+f"/{key}_i.inp", "w") as f:
            f.write(job_str)
        with open(f"{job_dir}/{_key}"+f"/{key}.sh", "w") as f:
            f.write(molpro_slurm_template.render(NAME=key))
        
for i,a in enumerate(cluster_flux["key"]):
    print(a)
    make_jobs(i)
```
